```python
import os
import pandas as pd

#Import csv files
data_csv1 = os.path.join(".", "election_data_1.csv")
data_csv2 = os.path.join(".", "election_data_2.csv")

#Read the imported csv files
data_csv1_pd = pd.read_csv(data_csv1)
data_csv1_pd.head()

data_csv2_pd = pd.read_csv(data_csv2)
data_csv2_pd.head()

#Merge two imported csv files and assign new to data_file
data_file = pd.concat([data_csv1_pd,data_csv2_pd])
data_file
```

```python
#Calculate Total Number of Votes
Total_Votes = len(data_file)
Total_Votes
```

```python
#Sum Votes per Candidate
Candidate_Total = data_file["Candidate"].value_counts()
Candidate_Total
```

```python
#Calculate the Vote Percentage per Candidate
Percentage = Candidate_Total / Total_Votes
Percentage
#Format as Percentage
Percentage_Format = Percentage * 100
Percentage_Format
```

```python
#Find the Candidate with most votes
Winner = Candidate_Total.head(1)
Winner
```

```python
#Print out results
print("Election Results")
print("----------------------------------------")
print("Total Votes " + str(Total_Votes))
print("----------------------------------------")
print(Percentage_Format)
print("----------------------------------------")
print("Winner " + str(Winner)) 
print("----------------------------------------")
```
